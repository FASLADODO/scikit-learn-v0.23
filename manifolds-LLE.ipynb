## [Locally Linear Embedding](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.LocallyLinearEmbedding.html#sklearn.manifold.LocallyLinearEmbedding)
- Seeks a low-D projection of the data which *preserves distances within local neighborhoods*. 
- Three stage algorithm:
    - 1) Nearest neighbor search (see isomap)
    - 2) Weight matrix construction - involves finding a solution for a _kxk_ linear equation, for each of _N_ local neighborhoods.
    - 3) Partial eigenvalue decomposition (see isomap)
- Overall complexity: $O[D \log(k) N \log(N)] + O[D N k^3] + O[d N^2]$
